```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
iris = load_iris()
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import scale

svc = SVC(kernel='sigmoid')
```

```python
svc = SVC(C=100.0)
```

```python
X = iris.data[:,0:4:3]
y = iris.target


plt.scatter(X.T[0],X.T[1],c=y)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='poly',degree=3)

X = scale(X)
```

```python
from sklearn.svm import SVC
svc = SVC()

svc.fit(X,y)
```

```python
svc.predict([8.0,2.0])
```

```python
xx,yy = np.meshgrid(np.linspace(-3,3.0,201),np.linspace(-2.0,2.0,201))

pred = svc.predict(np.c_[xx.ravel(),yy.ravel()])
Z_svc = pred.reshape(np.shape(xx))

plt.figure(figsize=(14,7))

plt.subplot(121)
plt.contourf(xx,yy,Z_svc,alpha=0.4)
plt.scatter(X.T[0],X.T[1],c=y)
plt.axis('scaled')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('svc',fontsize=20)
plt.xlim(-3.0,3.0)
plt.ylim(-2.0,2.0)
```
