```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
from sklearn.svm import LinearSVC

lvsc = LinearSVC(C=100.0)
```

```python

X = iris.data[(iris.target==1)|(iris.target==2),1:4:2]
y = iris.target[(iris.target==1)|(iris.target==2)]
```

```python
plt.scatter(X.T[0],X.T[1],c=y,cmap='Spectral_r')
```

```python
X.shape
```

```python
from sklearn.preprocessing import scale
X = scale(X)
```

```python
from sklearn.svm import LinearSVC
lsvc = LinearSVC(dual=False, C=100.0, penalty = 'l2')

lsvc.fit(X,y)
```

```python
lsvc.predict([8.0,2.0])
```

```python
xx,yy = np.meshgrid(np.linspace(-3,3.0,201),np.linspace(-2.0,2.0,201))

pred = lsvc.predict(np.c_[xx.ravel(),yy.ravel()])
Z_svmp = pred.reshape(np.shape(xx))

plt.figure(figsize=(14,7))

plt.subplot(121)
plt.contourf(xx,yy,Z_svmp,alpha=0.4)
plt.scatter(X.T[0],X.T[1],c=y)
plt.axis('scaled')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('svm',fontsize=20)
plt.xlim(-3.0,3.0)
plt.ylim(-2.0,2.0)
```
