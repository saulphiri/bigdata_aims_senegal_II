```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
data = iris.data
target =iris.target
```

```python
plt.scatter(data[:,1],data[:,2], cmap='brg')
```

```python
from sklearn.cluster import KMeans
kms = KMeans(n_clusters=5)
#kms.fit(data[:,1:3:1])
#y_pred = kms.predict(data[:,1:3:1])
y_pred = kms.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred, cmap='brg')
```

```python
kms.predict([[3.0,2.0],[4.0,5.0]])

```

MEANSHIFT

```python
from sklearn.cluster import MeanShift
msh = MeanShift(bandwidth=1.0)
#kms.fit(data[:,1:3:1])
#y_pred = kms.predict(data[:,1:3:1])
y_pred = msh.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred, cmap='brg')
```

AFFINITY PROPAGATION

```python
from sklearn.cluster import AffinityPropagation

afp = AffinityPropagation()
#afp.fit_predict([:,1:3:1])
y_pred = afp.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred, cmap='brg')
```

```python
np.unique(afp.labels_).size
```

AGGLOMERATIVE CLUSTERING

```python
from sklearn.cluster import AgglomerativeClustering

agc = AgglomerativeClustering(n_clusters=2, affinity='euclidean',linkage='average')
y_pred = agc.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred, cmap='brg')
```

```python
np.unique(afp.labels_).size
```

DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbs =DBSCAN(eps=0.5,min_samples=5)
y_pred =dbs.fit_predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1],data[:,2],c=y_pred, cmap='brg')
```
